# Question 5 (5pts)

How do you think the performance of your model varies in the train and test set as you increase(decrease) the number of examples in the training dataset? Explain why does it change in a particular way.

## Answer

As you increase the number examples it becomes more representative of the actual population and thus becomes more normally distributed. The more normally distrubted the data is, the more similar the train and test set will be. Thus, the performance between train and test will be closer. The opposite is true when there are less observations. If there are very few observations than the less similar the train and test sets will be and thus there will be worse performance going from training to the test set.
