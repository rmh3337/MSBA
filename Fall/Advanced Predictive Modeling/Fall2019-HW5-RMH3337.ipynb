P(D = 1) = 0.1   
P(D = 0) = 0.9  

P(Z = 1 | D = 1) = 0.7   
P(Z = 0 | D = 1) = 0.3  
P(Z = 1 | D = 0) = 0.8  
P(Z = 0 | D = 0) = 0.2  

P(T2 = 1 | D = 1) = 0.7           
P(T2 = 0 | D = 1) = 0.3         
P(T2 = 1 | D = 0) = 0.5     
P(T2 = 0 | D = 0) = 0.5

P(T1 = 0 | D = 1, Z = 0) = 0.2

P(T1 = 0 | D = 0, Z = 1) = 0.4

P(T1 = 0 | D = 0, Z = 0) = 0.9            


a.(4pts) What is the probability of having disease D(D=1) and getting a positive result on test T1(T1=1)?   
b.(3pts) What is the probability of not having disease D(D=0) and getting a positive result on test T1(T1=1)?   
c.(3pts) What is the probability of having disease D given a positive result on test T1?   


#### Answer:



### Q1.a

P(D=1,T1=1) = P(D=1) P(Z|D) P(T1|Z,D) 

             = P(D=1)P(Z=1|D=1)P(T1=1|Z=1,D=1) + P(D=1)P(Z=0|D=1)P(T1=1|Z=0,D=1)

            = 0.1*0.7*0.9 +0.1*0.3*0.8
            
            = 0.087

### Q1.b

P(D=0,T1=1)= P(D=0) P(Z|D) P(T1|Z,D)

            = P(D=0)P(Z=1|D=0)P(T1=1|Z=1,D=0) + P(D=0)P(Z=0|D=0)P(T1=1|Z=0,D=0)
            
            = 0.9*0.8*0.6 +0.9*0.2*0.1
            
            = 0.45


### Q1.c 

 P(D=1|T1=1)=P(D=1,T1=1)/P(T1=1)
 
           =P(D=1,T1=1)/(P(D=1,T1=1)+P(D=0,T1=1))
           
           =0.087/(0.087+0.45)
           
           =0.162

#### (1)

Build K-Nearest Neighbours (KNN) models with different values of k ranging from 2 to 50. Plot how the train and validation AUROC varies with k. Report the best AUROC on the validation set and the corresponding value of k. (10 pts)

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

auroc_train = []
auroc_val = []
k = []
for i in range(2,51):
    
    knn = Pipeline(steps=[('preprocessor', all_feature_preprocessor),
                      ('classifier', KNeighborsClassifier(n_neighbors=i))])
    
    knn.fit(X_train, y_train)
    pred_i_train = knn.predict_proba(X_train)[:,1]
    pred_i_val = knn.predict_proba(X_val)[:,1]
    auroc_train.append(roc_auc_score(y_train, pred_i_train))
    auroc_val.append(roc_auc_score(y_val, pred_i_val))
    k.append(i)
    
```

```python
plt.plot(auroc_train)
plt.plot(auroc_val)

print('Best AUROC on Validation set: ', max(auroc_val))
print('KNN that corresponds to the best AUROC on Validation set: ', k[auroc_val.index(max(auroc_val))])
```

#### (2)

We will reuse the Naive Bayes and Logistic regression models from the previous homework.

Ensemble the KNN model which obtains the best AUROC on the validation set with a Naive Bayes and Logistic regression model (with parameter class_weight='balanced') by averaging the posterior probabilities. Report the validation AUROC and F1 score. (5 pts)

```python
estimators = [('Best KNN', KNeighborsClassifier(n_neighbors=k[auroc_val.index(max(auroc_val))])),
                 ('Naive Bayes', GaussianNB()),
                 ('Logistic Regression', LogisticRegression(class_weight='balanced'))]

ensemble = VotingClassifier(estimators, voting='soft')

ens = Pipeline(steps=[('preprocessor', all_feature_preprocessor),
                      ('classifiers', ensemble)])

#fit model to training data
ens.fit(X_train, y_train)
#test our model on the test data
y_pred_proba = ens.predict_proba(X_val)[:,1]
y_pred = ens.predict(X_val)
auroc = roc_auc_score(y_val, y_pred_proba)
f1 = f1_score(y_val, y_pred)

print('F1 score:', f1)
print('AUROC: ', auroc)
```

```python
estimators = [('Best KNN', KNeighborsClassifier(n_neighbors=k[auroc_val.index(max(auroc_val))])),
                 ('Naive Bayes', GaussianNB()),
                 ('Logistic Regression', LogisticRegression(class_weight='balanced'))]

ensemble = VotingClassifier(estimators, voting='hard')

ens = Pipeline(steps=[('preprocessor', all_feature_preprocessor),
                      ('classifiers', ensemble)])

#fit model to training data
ens.fit(X_train, y_train)
#test our model on the test data
y_pred = ens.predict(X_val)
f1 = f1_score(y_val, y_pred)

print('F1 score:', f1)
```
